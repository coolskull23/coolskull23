```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

datasets = {
    "SO2TONS": "https://raw.githubusercontent.com/apownukepcc/ForecastingDailyEmissions/refs/heads/main/SO2TONS_dataset.csv",
    "NOXTONS": "https://raw.githubusercontent.com/apownukepcc/ForecastingDailyEmissions/refs/heads/main/NOXTONS_dataset.csv",
    "COTONS": "https://raw.githubusercontent.com/apownukepcc/ForecastingDailyEmissions/refs/heads/main/COTONS_dataset.csv"
}

data = pd.read_csv(datasets['SO2TONS'])

X = data['tavg'].values.reshape(-1, 1)  # Independent variable (Value)
y = data['Emissions_Load'].values  # Dependent variable (Emissions_Load)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', alpha=0.5)

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

plt.plot(X, y_pred, color='red', linewidth=2)

plt.xlabel('tavg')
plt.ylabel('Emissions_Load')
plt.title('SO2TONS Relationship between tavg and Emissions Load')

plt.show()

# The plot and the linear regression line indicate the relationship between 'tavg' (average temperature) and 'Emissions_Load' (SO2 emissions load).
# If the slope of the regression line is positive, it suggests that higher temperatures correlate with increased emissions load. Conversely, a negative slope would indicate a decrease in emissions load as temperature increases.
```
